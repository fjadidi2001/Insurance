```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
labels = ['Training Set', 'Testing Set', 'Validation Set']
sizes = [136222, 29191, 29191]
colors = sns.color_palette("coolwarm", len(sizes))

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}, shadow=True)

# Title
plt.title("Dataset Split Distribution", fontsize=14, fontweight='bold')

# Save plot to file
plt.savefig("dataset_distribution.png", dpi=128)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the original and resampled distribution
labels = ['Not Risky (0)', 'Risky (1)']
original_counts = [97302, 2698]
resampled_counts = [97302, 97302]

# Set the bar width
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bars
bars1 = ax.bar(r1, original_counts, color='skyblue', width=bar_width, edgecolor='grey', label='Original Distribution')
bars2 = ax.bar(r2, resampled_counts, color='salmon', width=bar_width, edgecolor='grey', label='Resampled Distribution')

# Add labels and title
ax.set_xlabel('Class', fontsize=14)
ax.set_ylabel('Number of Samples', fontsize=14)
ax.set_title('Comparison of Original and Resampled Class Distributions', fontsize=16)
ax.set_xticks([r + bar_width / 2 for r in range(len(labels))])
ax.set_xticklabels(labels)

# Add annotations to the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:,}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

# Add legend
ax.legend()

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
