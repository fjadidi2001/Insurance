```python
from graphviz import Digraph
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# رسم معماری مدل پیشنهادی با Graphviz
def draw_model_architecture():
    dot = Digraph(format='png')
    dot.attr(size='8,8')

    dot.node('Input', 'داده‌های تله‌ماتیک و بیمه‌ای', shape='parallelogram', style='filled', fillcolor='lightblue')
    dot.node('TabNet', 'ویژگی‌های منتخب (TabNet)', shape='box', style='filled', fillcolor='lightyellow')
    dot.node('CNN', 'استخراج الگوهای پیچیده (CNN)', shape='box', style='filled', fillcolor='lightpink')
    dot.node('xLSTM', 'وابستگی‌های زمانی بلندمدت (xLSTM)', shape='box', style='filled', fillcolor='lightgreen')
    dot.node('Output', 'پیش‌بینی ریسک بیمه‌ای', shape='parallelogram', style='filled', fillcolor='lightblue')

    dot.edge('Input', 'TabNet')
    dot.edge('TabNet', 'CNN')
    dot.edge('TabNet', 'xLSTM')
    dot.edge('CNN', 'Output')
    dot.edge('xLSTM', 'Output')

    dot.render('model_architecture')

# نمودار مقایسه عملکرد مدل‌ها
def plot_performance_comparison():
    models = ['TabNet', 'CNN', 'xLSTM', 'Hybrid Model']
    metrics = ['Accuracy', 'F1-score', 'AUC', 'MCC']
    performance = np.random.uniform(0.75, 0.98, (4, 4))  # داده‌های تصادفی برای نمایش

    df_performance = pd.DataFrame(performance, index=models, columns=metrics)
    df_performance.plot(kind='bar', figsize=(10, 5), colormap='viridis', edgecolor='black')
    plt.title('مقایسه عملکرد مدل‌ها')
    plt.xlabel('مدل')
    plt.ylabel('مقدار')
    plt.xticks(rotation=0)
    plt.legend(title='معیارها')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# نمودار اهمیت ویژگی‌ها (Heatmap)
def plot_feature_importance():
    features = ['سرعت متوسط', 'تعداد ترمز ناگهانی', 'مسافت طی‌شده', 'شتاب‌گیری ناگهانی', 'زمان رانندگی شبانه']
    importance = np.random.uniform(0.1, 1, (5, 5))
    df_importance = pd.DataFrame(importance, index=features, columns=['TabNet', 'CNN', 'xLSTM', 'ترکیبی', 'SHAP'])

    plt.figure(figsize=(8, 6))
    sns.heatmap(df_importance, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('نقشه حرارتی اهمیت ویژگی‌ها در مدل‌های مختلف')
    plt.show()

# اجرای توابع
draw_model_architecture()
plot_performance_comparison()
plot_feature_importance()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the original and resampled distribution
labels = ['Not Risky (0)', 'Risky (1)']
original_counts = [97302, 2698]
resampled_counts = [97302, 97302]

# Set the bar width
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = np.arange(len(labels))
r2 = [x + bar_width for x in r1]

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bars
bars1 = ax.bar(r1, original_counts, color='skyblue', width=bar_width, edgecolor='grey', label='Original Distribution')
bars2 = ax.bar(r2, resampled_counts, color='salmon', width=bar_width, edgecolor='grey', label='Resampled Distribution')

# Add labels and title
ax.set_xlabel('Class', fontsize=14)
ax.set_ylabel('Number of Samples', fontsize=14)
ax.set_title('Comparison of Original and Resampled Class Distributions', fontsize=16)
ax.set_xticks([r + bar_width / 2 for r in range(len(labels))])
ax.set_xticklabels(labels)

# Add annotations to the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:,}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_value_labels(bars1)
add_value_labels(bars2)

# Add legend
ax.legend()

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
